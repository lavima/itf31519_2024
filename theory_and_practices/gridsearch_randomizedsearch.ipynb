```python
# Notebook used to create figures in slides on Hyperparameter Tuning

import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,5))
grid_x = np.arange(0, 11)
grid_y = np.arange(0, 11)
grid = np.meshgrid(grid_x, grid_y)

randint_x = np.random.randint(0,10, (8,8))
randint_y = np.random.randint(0,10, (8,8))
    
randgrid_x = np.random.uniform(0.0, 10.0, (10, 10))
randgrid_y = np.random.uniform(0.0, 10.0, (10, 10))

axs[0].scatter(grid[0].flatten(), grid[1].flatten())
axs[1].scatter(randint_x.flatten(), randint_y.flatten())
axs[2].scatter(randgrid_x.flatten(), randgrid_y.flatten())
```
